```python
sns.countplot(x="is_churned", data=df)
plt.title("Churn Distribution")
plt.savefig("../results/plots/churn_distribution.png")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="subscription_type", hue="is_churned", data=df)
plt.title("Churn vs Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Number of Users")
plt.savefig("../results/plots/churn_vs_subscription_type.png")
plt.show()
```

```python
sns.countplot(x="gender", hue="is_churned", data=df)
plt.title("Churn vs Gender")
plt.savefig("../results/plots/churn_vs_gender.png")
plt.show()
```

```python
sns.countplot(x="device_type", hue="is_churned", data=df)
plt.title("Churn vs Device Type")
plt.savefig("../results/plots/churn_vs_device_type.png")
plt.show()
```

```python
sns.countplot(x="offline_listening", hue="is_churned", data=df)
plt.title("Churn vs Offline Listening")
plt.savefig("../results/plots/churn_vs_offline_listening.png")
plt.show()
```

```python
sns.boxplot(x="is_churned", y="listening_time", data=df)
plt.title("Listening Time vs Churn")
plt.savefig("../results/plots/churn_vs_listening_time.png")
plt.xlabel("Churned (0 = No, 1 = Yes)")
plt.ylabel("Listening Time")
plt.show()
```

```python
sns.boxplot(x="is_churned", y="songs_played_per_day", data=df)
plt.title("Songs Played per Day vs Churn")
plt.savefig("../results/plots/churn_vs_songs_played.png")

plt.show()
```

```python
sns.boxplot(x="is_churned", y="skip_rate", data=df)
plt.title("Skip Rate vs Churn")
plt.savefig("../results/plots/churn_vs_skip_rate.png")
plt.show()
```

```python
sns.boxplot(x="is_churned", y="age", data=df)
plt.title("Age vs Churn")
plt.savefig("../results/plots/churn_vs_age.png")
plt.show()
```

```python
sns.boxplot(x="is_churned", y="ads_listened_per_week", data=df)
plt.title("Ads Listened per Week vs Churn")
plt.savefig("../results/plots/churn_vs_ads_listened_per_week.png")
plt.show()
```
